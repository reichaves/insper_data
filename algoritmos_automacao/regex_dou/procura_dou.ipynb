```python
# Funções para buscar CPFs, CPFs mascarados e CNPJs nos textos do DOU
```

```python
def get_cpfs(text):

    pattern = r"\d{3}\.\d{3}\.\d{3}-\d{2}"

    return re.findall(pattern, text)
```

```python
def get_cpfs_mascarado(text):

   pattern = r"\*\*\*\.\d{3}\.\d{3}-\*\*"

   return re.findall(pattern, text)
```

```python
def get_cnpjs(text):

    pattern = r"\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}"

    return re.findall(pattern, text)
```

```python
# Faz buscas de dois termos de interesse para posterior investigação - gera dicionário
```

```python
# inexigibilidade de licitação
# dispensa de licitação
entidades = []
```

```python
for num, row in df.iterrows():
    texto = row['paragraphs']
    match = re.search(r'inexigibilidade', texto)
    
    if match: 
        print("---------------")
        print('Start Index:', match.start())
        print('End Index:', match.end())
        print(texto)
        print("")
        print("CPFs encontrados : ", get_cpfs(texto))
        print("CPFs mascarados encontrados : ", get_cpfs_mascarado(texto))
        print("CNPJs encontrados : ", get_cnpjs(texto))
        
        dicionario = {"data_dou": "17/12/2021",
                      "secao": "3", 
                      "termo_procurado": "inexigibilidade",
                      "texto": texto,
                      "cpfs": get_cpfs(texto), 
                      "cpfs_mascarados": get_cpfs_mascarado(texto), 
                      "cnpjs": get_cnpjs(texto)
                          }
        entidades.append(dicionario)
        
        print("")
```

```python
# Busca por dispensa
```

```python
for num, row in df.iterrows():
    texto = row['paragraphs']
    match = re.search(r'dispensa', texto)
    if match: 
        print("---------------")
        print('Start Index:', match.start())
        print('End Index:', match.end())
        print(texto)
        print("")
        print("CPFs encontrados : ", get_cpfs(texto))
        print("CNPJs encontrados : ", get_cnpjs(texto))
        
        dicionario = {"data_dou": "17/12/2021",
                      "secao": "3", 
                      "termo_procurado": "dispensa",
                      "texto": texto,
                      "cpfs": get_cpfs(texto), 
                      "cpfs_mascarados": get_cpfs_mascarado(texto), 
                      "cnpjs": get_cnpjs(texto)
                          }
        entidades.append(dicionario)
        
        print("")
```
